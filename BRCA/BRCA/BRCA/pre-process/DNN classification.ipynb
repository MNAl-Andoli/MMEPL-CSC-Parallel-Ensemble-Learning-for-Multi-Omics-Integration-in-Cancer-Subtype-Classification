```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
data = np.loadtxt('latent_data.csv', delimiter=',')
print(data.shape)

labels = np.loadtxt('encoded_labels.csv')
print(labels.shape)
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
input_dim=X_train.shape[1]
num_classes=num_classes = len(np.unique(labels))

early_stop = EarlyStopping(monitor='loss', patience=2)
model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(input_dim,),kernel_initializer='uniform'))
#model.add(Dense(32, activation='relu',kernel_initializer='random_uniform'))
model.add(Dense(64, activation='relu',kernel_initializer='random_uniform'))
#model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
#model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
hist = model.fit(X_train , y_train , epochs=100)
```

```python
y_predict = model.predict(X_test)
y_predict = np.argmax(y_predict, axis=1)
```

```python
print(classification_report(y_predict,y_test), '\n')
```
